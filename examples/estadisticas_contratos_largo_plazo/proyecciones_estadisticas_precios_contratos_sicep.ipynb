### Filtrar la máxima versión para el periodo de consulta de los datos
El método filtra para el dataset de consulta, la máxima versión para cada fecha proyectada y cada agente comprador

```python
def max_version(df, columns_groupby, column_version, columns_data):
    """
    Función to get the last version of the data and unpivot it.
    """
    df[column_version] = pd.to_datetime(df[column_version])

    # Agrupar por las columnas especificadas y obtener la última versión de cada grupo
    df_last_version = df.loc[df.groupby(columns_groupby)[column_version].idxmax()]
    
    # Aplicar unpivot a las columnas de datos especificadas    
    df_last_version = pd.melt(
        df_last_version,
        id_vars=columns_groupby,  # Columnas a conservar como identificadores
        value_vars=columns_data,  # Columnas para aplicar unpivot
        var_name='Concept',  # Nombre de la nueva columna para los conceptos
        value_name='Value'   # Nombre de la nueva columna para los values
    )   
    
    return df_last_version
```

```python
# Visualizar los nombres de las columnas
df_dataset.columns
```

```python
df_last_version = max_version(
                            df=df_dataset,
                            columns_groupby=['FechaProyectada', 'CodigoAgenteComprador'], 
                            column_version='FechaPublicacion', 
                            columns_data=['PrecioMinimo', 'Percentil5', 'PPP', 'PPS', 'Mediana', 'Percentil95',
                                            'PrecioMaximo', 'LimiteInferiorIC95Ponderado', 'LimiteSuperiorIC95Ponderado',
                                            'LimiteInferiorIC95Simple', 'LimiteSuperiorIC95Simple', 'TotalContratos']
                            )

df_last_version.to_clipboard()
df_last_version.head(10)
```

```python
def generate_graph(df, group_axis, title, xlabel, ylabel, x_axis='FechaProyectada', y_axis='Value',):
    
    """
    Genera las gráficas de líneas de los datos del DataFrame.
    df: DataFrame con los datos a graficar.
    group_axis: Columna por la cual se agruparán los datos (e.g., 'Concept').
    title: Título de la gráfica.
    xlabel: Etiqueta del eje X.
    label: Etiqueta del eje Y.
    x_axis: Columna del eje X (default: 'FechaProyectada').
    y_axis: Columna del eje Y (default: 'Value').    
    """
    
    fig, ax = plt.subplots(figsize=(12, 4))  
    
    # Agrupar por el eje especificado y graficar cada grupo
    for concept, group in df.groupby(group_axis):
        ax.plot(group[x_axis], group[y_axis], marker= '.', label=concept)

    # Configurar título y etiquetas de los ejes
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    # Configurar los ejes X e Y
    ticks_to_use = df[x_axis][::12]  # Adjust the step (e.g., every 5th date)
    ax.set_xticks(ticks_to_use)
    ax.set_xticklabels(ticks_to_use, rotation=45)
    
    # Establecer límites del eje Y para una mejor visualización
    min_value, max_value = df[y_axis].min(), df[y_axis].max()
    ax.set_ylim(min_value*0.5, max_value*1.5)
    
    plt.grid(True)
    ax.legend(title=group_axis)

    plt.show()
```

```python
def box_plot(df, group_by, concepts, x_label, y_label):
    
    """
    Genera un boxplot de los datos del DataFrame agrupados por una columna específica.
    df: DataFrame con los datos a graficar.
    group_by: Columna por la cual se agruparán los datos (e.g., 'CodigoAgenteComprador').
    concepts: Lista de conceptos a incluir en el boxplot (e.g., ['PPP', 'PPS']).
    x_label: Etiqueta del eje X.
    y_label: Etiqueta del eje Y.
    """
    
    # Agrupar los datos por la columna especificada y filtrar la columna 'Value' por los conceptos seleccionados
    grouped_data = [df[(df[group_by] == x) & (df.Concept.isin(concepts))]['Value'] for x  in df[group_by].unique()]

    # Crear el boxplot
    plt.figure(figsize=(12, 6))
    plt.boxplot(grouped_data, labels=df[group_by].unique(), patch_artist=True)


    # Ajustar el estilo del boxplot
    plt.title(f'Distribución de {concepts} por {group_by}')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.show()
```

```python
# Graficar un boxplot de precios máximos por agente comprador
concepts = ['PrecioMaximo']
agents = ['EPMC', 'ESSC', 'ENDC', 'CASC', 'HLAC', 'CNSC', 'CQTC', 'EBSC' ]
group_by = 'CodigoAgenteComprador'

# Filtrar el DataFrame para los agentes seleccionados
df_agente = df_last_version[(df_last_version.CodigoAgenteComprador.isin(agents))]
box_plot(df_agente, group_by, concepts, x_label='Agente Comprador', y_label='Precio COP/kWh')
```
